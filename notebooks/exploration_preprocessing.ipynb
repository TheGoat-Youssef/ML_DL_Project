Visualisation aléatoire

```python
import random
import cv2

for cls in config.CLASS_NAMES:
    images = list((config.RAW_DIR/"train"/cls).glob("*"))
    plt.figure(figsize=(10,2))
    plt.suptitle(cls)

    for i in range(5):
        img = cv2.imread(str(random.choice(images)))
        plt.subplot(1,5,i+1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
    plt.show()
```
